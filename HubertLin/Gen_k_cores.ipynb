## Genrate k-cores feature

k-core is kind of graphical feature pointing out the relation between data and how "popular" a data is.

```python
import numpy as np
import pandas as pd
```

```python
df_kscore = pd.read_csv('./question_kcores.csv')
```

```python
df_train = pd.read_csv('../dataset/quora-question-pairs/train.csv')
df_test = pd.read_csv('../dataset/quora-question-pairs/test.csv')

df_train['question1'] = df_train['question1'].str.lower()
df_train['question2'] = df_train['question2'].str.lower()
df_test['question1'] = df_train['question1'].str.lower()
df_test['question2'] = df_train['question2'].str.lower()
```

```python
d = {}

for i in range(len(df_kscore)):
    series = df_kscore.ix[i]
    d[series['question']] = series['kcores']
    if i%100000==0:
        print(i)
```

```python
SET = 'train'
# SET = 'test'

if SET=='train':
    df = df_train
else:
    df = df_test

out1 = []
out2 = []
for i,series in df.iterrows():
    
    q1 = d[ series['question1'] ]
    q2 = d[ series['question2'] ]
    
    out1.append(q1)
    out2.append(q2)
    
    if i%100000==0:
        print(i, '/', len(df))
    
df_res = pd.DataFrame(np.array([out1,out2]).reshape(-1,2), columns=['kscore1','kscore2'])
df_res.to_csv('./features_from_model/'+SET+'/k_scrore.csv', index=False)
```
